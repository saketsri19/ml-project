```python
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
for i in ['Price', 'Open', 'High', 'Low', 'Vol.']:
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
```
